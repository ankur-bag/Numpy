Numpy Indexing and Slicing of Vectors

```python
import numpy as np
```

```python
arr = np.arange(11,21)
arr
```

```python
arr[5] #indexing
```

```python
arr[1::6] #slicing
```

Numpy indexing and slicing of Matrix

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]
```

```python
arr[4,4]
```

```python
arr[0:2,1:3]
```

```python
arr[3:6,3:5]
```

```python
arr[:,2] #slicing out any column
```

Bollean Indexing

```python
arr = np.arange(11,21)
arr
```

```python
bool_index = arr%2 ==0
bool_index
```

```python
arr = arr[bool_index]
arr
```
