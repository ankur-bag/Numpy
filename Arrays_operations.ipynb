Arithmetic Operations

```python
import numpy as np
```

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
a1+a2
```

```python
a1-a2
```

```python
a1*a2
```

```python
a1/a2
```

```python
a1//a2 #floor division
```

```python
l = [10,20,30,40]
arr = np.array(l)

arr
```

```python
for i in l:
    print(i*10)
```

```python
arr+10
```

Deep and Shallow copy

```python
a = np.arange(1,21)
a
```

```python
slice =a[:5]
slice = slice*10
slice
```

```python
b=a.copy() 
b[0] =90
b
```

Matrix Operations

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
```

```python
A
```

```python
B
```

```python
A @ B #actual multiplication of Matrix
```

```python
np.dot(A,B)
```

Stacking of matrix

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a,b)) # vertical stacking
```

```python
np.hstack((a,b)) # horizontal stack
```

```python
np.column_stack((a,b))
```
